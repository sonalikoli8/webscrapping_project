```python
import  pandas as pd
d={'title':[],'author':[],'price':[]}
for i in range(1,35):
    url=f"https://www.bookdepository.com/bestsellers?page={i}"
    page=requests.get(url)
    Soup=BeautifulSoup(page.content,'html5lib')
    data=Soup.find_all(class_="book-item")
    for f in data:
        d['title'].append(f.find(class_='title').get_text().strip())
        d['author'].append(f.find(class_="author").get_text().strip())
        d['price'].append(f.find(class_="price-wrap").get_text().strip().split('\n')[0].lstrip('â¹'))
df=pd.DataFrame(d)
df.head(30)
```
